# Primer Ejemplo
## Top de las 100 videos mas vistos de acuerdo con el numero de views


```python
top100_videos_mas_vistos = youtube_dataset['views'].sort_values(ascending=False).head(100)
videos_mas_vistos = youtube_dataset.loc[top100_videos_mas_vistos.index,['video_id','trending_date','title','channel_title','category_id','tags','views','likes','dislikes','comment_count']]
print('TOP 100:')
display(videos_mas_vistos)
```

# Segundo Ejemplo

# Noveno Ejemplo
## Histograma

```python
top100_videos_trending_date = youtube_dataset['category_id'].sort_values(ascending=False).head(10000)
videos_mas_trending = youtube_dataset.loc[top100_videos_trending_date.index,['category_id','title','comment_count','likes']]
print('TOP 10000:')
display(videos_mas_trending)

```

```python

data = videos_mas_trending['category_id'].values
print(data.shape)
(n, bin_edges) = np.histogram(data, bins=50)
plt.plot(bin_edges[1:],n)
plt.show()
```

```python
data = videos_mas_comentados['comment_count'].values
print(data.shape)
(n, bin_edges) = np.histogram(data, bins=50)
plt.plot(bin_edges[1:],n)
plt.show()
```
